```python
species.fillna('No Concern', inplace=True)
status = species.groupby('conservation_status').count().reset_index()
null_spec = species['conservation_status'].value_counts(dropna=False)
print(null_spec)
print('\n\n')
print(status)
```

```python
different_parks = observations['park_name'].unique()
print(f'The different parks observed are: {[x for x in different_parks]}, there are {len(different_parks)} in total.')
```

```python
animal_categories = species['category'].unique()
print(f'The different categories of animals are {animal_categories}')
```

```python
endangered_species = species[species['conservation_status'] == 'Endangered']
concern_species = species[species['conservation_status'] == 'Species of Concern']
threatened_species = species[species['conservation_status'] == 'Threatened']
recovering_species = species[species['conservation_status'] == 'In Recovery']
```

```python
conservation_category = species[species['conservation_status'] != 'No Concern'].groupby(['conservation_status', 'category']).scientific_name.count().unstack()
conservation_category
```

```python
ax = conservation_category.plot(kind = 'bar', stacked=True, figsize=(15,10))
ax.set_ylabel('Number of Species')
ax.set_xlabel('Conservation Status')
plt.grid()
plt.show()
```

```python
species['is_protected'] = (species['conservation_status'] != 'No Concern')
```

```python
category_counts = species.groupby(['category', 'is_protected'])\
    .scientific_name.nunique()\
    .reset_index()\
    .pivot(index='category', columns='is_protected', values='scientific_name')\
    .reset_index()
category_counts.columns = ['category', 'not_protected', 'protected']
(category_counts)
```

```python
category_counts['percent_protected'] = category_counts['protected']/\
    (category_counts['protected'] + category_counts['not_protected']) * 100

print(category_counts)
```

```python
# Chi2 Contingencies

mam_bir = [[30, 146],
          [75, 413]]

bir_rep = [[75, 413],
          [3, 75]]

mam_rep = [[40, 146],
          [3, 75]]

```

```python
from scipy.stats import chi2_contingency
```

```python
(chi2_contingency(mam_bir))
```

```python
chi2_contingency(mam_rep)
```

```python
chi2_contingency(bir_rep)
```

The most common use of the chi square test in statistical practice is to assess whether one variable is indpendent of another, for instance, the p value for the chi square test between mammals and birds comes in very high at 0.69, suggesting that these two variables are very independent of one another nad dont have much of a direct correlation with each other.  Mammals and reptiles on the other hand seem to be much more connected to each other, with a p value rounded up to 0.001, the connection between these two species appears to be very significant,

```python
from itertools import chain
import string

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

common_Names = species[species.category == "Mammal"]\
    .common_names\
    .apply(remove_punctuations)\
    .str.split().tolist()

common_Names[:6]
```

```python
cleanRows = []

for item in common_Names:
    item = list(dict.fromkeys(item))
    cleanRows.append(item)
    
cleanRows[:6]
```

```python
res = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in cleanRows))
res[:6]
```

```python
words_counted = []

for i in res:
    x = res.count(i)
    words_counted.append((i,x))

pd.DataFrame(set(words_counted), columns =['Word', 'Count']).sort_values("Count", ascending = False).head(10)
```

```python
species['is_bat'] = species.common_names.str.contains(r"\bBat\b", regex = True)

species.head(10)
```

```python
species[species.is_bat]
```

```python
bat_observations = observations.merge(species[species.is_bat])
bat_observations
```

```python
bat_observations.groupby('park_name').observations.sum().reset_index()
```

```python
obs_by_park = bat_observations.groupby(['park_name', 'is_protected']).observations.sum().reset_index()
obs_by_park
```

```python
plt.figure(figsize=(16, 4))
sns.barplot(x=obs_by_park.park_name, y= obs_by_park.observations, hue=obs_by_park.is_protected)
plt.xlabel('National Parks')
plt.ylabel('Number of Observations')
plt.title('Observations of Bats per Week')
plt.show()
```
